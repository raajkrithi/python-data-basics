```python
import pandas as pd
import matplotlib.pyplot as plt

url ="https://raw.githubusercontent.com/mwaskom/seaborn-data/master/tips.csv"

df = pd.read_csv(url)
df.head()
```

Total Bill vs tip

```python
plt.scatter(df["total_bill"], df["tip"])
plt.xlabel("Total Bill")
plt.ylabel("Tip")
plt.title("Total bill vs Tip")
plt.show()
```

Average tip by day

```python
df.groupby("day")["tip"].mean().plot(kind="bar")
plt.title("Average Tip by Day")
plt.ylabel("Tip amount")
plt.show()
```

Insights from Restaurant Tips Dataset

- Tips increase as total bill increases -> positive relationship
- Weekend day Saturday and Sunday shows higher average tips
- Indicates customer spending behavior affects tipping

```python
df.groupby("day")["tip"].mean().to_csv("avg_tip_by_day.csv")
```

Histogram view

```python
plt.hist(df["total_bill"], bins=20)
plt.title("Total Bill Distribution")
plt.xlabel("Bill Amount ($)")
plt.ylabel("Number of Customers")
plt.show()
```
